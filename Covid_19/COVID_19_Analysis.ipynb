## Statewise  Figures

```python
data_recent = data_recent.sort_values(by=['Confirmed'], ascending= False)
plt.figure(figsize =(14,8), dpi=100)
plt.barh(data_recent['State/UnionTerritory'][:10], data_recent['Confirmed'][:10],
         align='center', color='#D32F2F')
plt.xticks([0,1000000,2000000,3000000,4000000,5000000,6000000],['0','1M','2M','3M','4M','5M','6M'])
plt.xlabel('Number of Confirmed Cases', size = 14)
plt.title('States with maximum confirmed cases [Top 10]', size = 18)
plt.show()
```

Maharashtra has more than 6 million confirmed cases, highest in India. Wheras Kerala and Karnataka has around 3 million reported cases, Tamil Nadu has approximately 2.5 million reported cases while the rest have less than 2 million reported cases.

```python
data_recent = data_recent.sort_values(by=['Deaths'], ascending= False)
plt.figure(figsize =(14,8), dpi=100)
plt.barh(data_recent['State/UnionTerritory'][:10], data_recent['Deaths'][:10],
         align='center', color='#D32F2F')
plt.xticks([0,20000,40000,60000,80000,100000,120000],['0','20K','40K','60K','80K','10K','12K'])
plt.xlabel('Number of Deaths', size = 14)
plt.title('States with maximum deaths[Top 10]', size = 18)
plt.show()
```

```python
data_recent = data_recent.sort_values(by=['Cured'], ascending= False)
plt.figure(figsize =(14,8), dpi=100)
plt.barh(data_recent['State/UnionTerritory'][:10], data_recent['Cured'][:10],
         align='center', color='#388E3C')
plt.xticks([0,1000000,2000000,3000000,4000000,5000000,6000000],['0','1M','2M','3M','4M','5M','6M'])
plt.xlabel('No. of Recoveries', size = 14)
plt.title('States with maximum recoveries [Top 10]', size = 18)
plt.show()
```

The states which contributes most to the number of cases in India also have most number of recoveries which means recovering from Covid-19 has very high chances if proper treatment is followed.

But this doesn't gives us idea about the state of healthcare in different states, so to get an idea we should check Death rate(Death per confirmed case) of states.

## Death Rate Calculation

```python
data_recent['Death Rate'] = (data_recent['Deaths'] / data_recent['Confirmed'])
data_recent.isna().sum()
data_recent
```

```python
data_recent['Recovery Rate'] = (data_recent['Cured'] / data_recent['Confirmed'])
data_recent
```

```python
data_recent = data_recent.sort_values(by = ['Death Rate'], ascending=False)
plt.figure(figsize =(14,8), dpi=100)
plt.barh(data_recent['State/UnionTerritory'][:10], data_recent['Death Rate'][:10],
         align='center', color='#D32F2F')
plt.xlabel('Death Rate', size = 14)
plt.title('States with maximum Death Rate [Top 10]', size = 18)
plt.show()
```

```python
data_recent = data_recent.sort_values(by = ['Recovery Rate'], ascending = False)
plt.figure(figsize=(12,8), dpi=80)
plt.barh(data_recent['State/UnionTerritory'][:10], data_recent['Recovery Rate'][:10], 
        align='center',color='#388E3C')
plt.xlabel('Recovery Rate', size = 14)
plt.xlim(0.985,1)
plt.title('States with maximum Recovery Rate[Top 10]', size = 18)
plt.show()
```
